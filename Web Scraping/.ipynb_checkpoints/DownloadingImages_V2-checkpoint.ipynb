# Obtener las im√°genes del sitio pngmart.com

```python
import requests
from bs4 import BeautifulSoup
```

```python
site_map = 'https://www.pngmart.com/sitemap.xml'
```

```python
response = requests.get(site_map)
```

```python
xml = response.text
```

```python
soup = BeautifulSoup(xml, 'xml')
```

```python
soup.find_all('loc')[1]
```

```python
for url1 in soup.find_all('loc'):
    print("url1: "+url1.text)
    url2 = requests.get(url1.text)
    xml2 = url2.text
    soup2 = BeautifulSoup(xml2,'xml')
    for image in soup2.find_all('image:loc'):
        print("image: "+image.text)
```
