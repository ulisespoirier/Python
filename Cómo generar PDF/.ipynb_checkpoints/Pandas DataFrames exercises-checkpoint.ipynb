## Arreglos booleanos (Mask):

### Dado el DataFrame de marvel_df pandas, haz una m√°scara que muestre los personajes femeninos:

```python
marvel_df[marvel_df['G√©nero'] == 'female']
```

```python
mask = marvel_df['G√©nero'] == 'female'
marvel_df[mask]
```

### Dado el DataFrame de marvel df pandas, obtenga los personajes con creados despu√©s de 1970:

```python
marvel_df[marvel_df['A√±o'] > 1970]
```

### Dado el DataFrame de marvel df pandas, obtenga los personajes femeninos creados despu√©s de 1970:

```python
mask = (marvel_df['G√©nero'] == 'female') & (marvel_df['A√±o'] > 1970)
marvel_df[mask]
```
